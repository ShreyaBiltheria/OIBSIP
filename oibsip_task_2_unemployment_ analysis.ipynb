## Correlation Matrix for first dataset 

```python
#calculating the correlation coefficients between all pairs of numerical columns in the DataFrame df1
df1.corr()
```

```python
#Correlation Matrix for df1
sns.heatmap(df1.corr(),cmap='viridis',annot=True)
plt.show()
```

1.The 'Estimated Unemployment Rate (%)' demonstrates a weak negative correlation (-0.22) with 'Estimated Employed,' suggesting that as unemployment increases slightly, the number of employed individuals tends to decrease marginally.

2.Both 'Estimated Unemployment Rate (%)' and 'Estimated Labour Participation Rate (%)' exhibit very weak correlations (close to 0) with each other, indicating an almost negligible linear relationship between the unemployment rate and labor participation rate.

3.There's no substantial linear association between the estimated unemployment rate and the estimated labor participation rate or the number of employed individuals.

## Correlation Matrix for second dataset

```python
#calculating the correlation coefficients between all pairs of numerical columns in the DataFrame df2
df2.corr()
```

```python
#Correlation Matrix for df1
sns.heatmap(df2.corr(),cmap='magma',annot=True)
plt.show()
```

1.Estimated Unemployment Rate' and 'Estimated Employed' display a moderate negative correlation (-0.25), indicating a decrease in employment as unemployment rises.

2.'Estimated Labour Participation Rate' shows negligible correlations with unemployment and employment variables (-0.07 to -0.05), signifying a weak relationship.

3.The 'latitude' columns have weak associations with other variables, with the second 'latitude' column demonstrating a slightly stronger connection (0.13) with the 'Estimated Labour Participation Rate.'

4.In essence, there's a notable negative relationship between unemployment and employment rates, while the relationship between latitude and labor participation seems slightly more pronounced compared to other associations within the dataset.

### Histplot for df1

```python
# For df1
plt.figure(figsize=(10,6))
sns.histplot(df1[' Estimated Unemployment Rate (%)'], bins=20, kde=True, color='purple', label='df1')
plt.title('Distribution of Estimated Unemployment Rates - df1')
plt.xlabel('Estimated Unemployment Rate')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

## Barplot for df1

```python
# bar plot of Estimated Unemployment Rate by Region
plt.bar(df1['Region'],df1[' Estimated Unemployment Rate (%)'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Unemployement Rate')
plt.title('Unemployement V/S State')
plt.show()
```

Clearly, from the above barplot we can say that Puducherry has highest Unemployment rate followed by Jharkhand and then Bihar.

```python
# bar plot of Estimated Employed by Region
plt.figure(figsize=(8,8))
sns.barplot(y="Region",x=" Estimated Employed",data=df1)
plt.show()
```

From above barplot we can say that Uttar Pradesh had highest estimated employed rate during the that period.

```python
# bar plot of Average Unemployment Rate by Region
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df1)
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Average Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

In above barplot we can see that Tripura and Haryana have highest average estimated unemployment rate(%).

## DRAWING INSIGHTS

Estimated Unemployment Rate (%): The percentage of individuals unemployed in each state of India.
    
Estimated Employed Individuals: The count of people currently employed.
    
Estimated Labour Participation Rate (%): The proportion of the working population (age group: 16-64 years) participating in the labor force, either employed or actively seeking employment.

In both the datasets most estimated employed were in Uttar Pradesh and Maharashtra.

The states/territories with highest unemployment rates considering both the datasets are:-
    
Tripura,Haryana,Bihar,Puducherry,Jharkhand,Tamil Nadu,Jammu & Kashmir and Delhi.

The most affected states/territories in India during the lockdown in case of unemployment were:

Tripura,Haryana,Bihar,Puducherry,Jharkhand,Jammu & Kashmir and Delhi.

### INSIGHT SUMMARY

Insight Summary:

1.Unemployment, quantified by the unemployment rate, reflects the percentage of individuals without employment within the total labor force. Amid the COVID-19 pandemic, there was a significant surge in the unemployment rate, prompting a need for comprehensive data analysis. 

2.In both datasets examined, Uttar Pradesh and Maharashtra consistently housed the highest estimated employed populations.

3.However, focusing on the unemployment rates in these datasets unveils the most affected states/territories during the pandemic-induced lockdown Tripura, Haryana, Bihar, Puducherry, Jharkhand, Jammu & Kashmir, and Delhi emerged as the regions experiencing the highest unemployment rates. 

4.These insights will help in  analyzing and understanding the fluctuations in unemployment rates, especially during periods of crisis like the COVID-19 pandemic.
