## SPLITING THE DATA 

```python
X = df.drop(['Car_Name','Selling_Price'],axis=1)
Y = df['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

## SPLITING THE TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

```python
from sklearn import metrics
```

```python
from sklearn.model_selection import cross_val_score
```

## MODEL TRAINING AND EVALUTION

## 1.Linear Regression

```python
from sklearn.linear_model import LinearRegression

# loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

```python
# prediction on Training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
#Visualizing the actual and predicted prices
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on Training data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
#visualizing actual prices and predicted prices
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
residuals = Y -lin_reg_model.predict(X)
sns.histplot(residuals, kde=True)
plt.title('Residuals Distribution')
plt.xlabel('Residuals')
plt.show()
```

```python
plt.scatter(lin_reg_model.predict(X), residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```

## CONCLUSION

In this car price prediction project, we embarked on a comprehensive journey through the key steps of machine learning model development. We initiated the process by collecting extensive data on various aspects, such as brand reputation, car features, horsepower, and mileage. Understanding the dataset provided crucial insights into patterns and relationships between different features, setting the foundation for effective model development.

Data preprocessing played a pivotal role in ensuring the quality and reliability of the model. Handling missing data, performing feature engineering, and encoding categorical variables were crucial steps in enhancing the predictive capabilities of the machine learning model. Normalizing or scaling numerical features further contributed to the robustness of the model.

The dataset was wisely split into training and testing sets, allowing for an accurate evaluation of the model's performance. This division is essential for assessing how well the model generalizes to new, unseen data.

Two regression algorithms, namely Linear Regression and Lasso Regression, were employed for model development. The evaluation metrics, including R-squared Error on both training and test data, provided valuable insights into the performance of each model.

For the Linear Regression model, the R-squared Error on the training data was 0.8796, indicating a strong fit. The R-squared Error on the test data was 0.8366, suggesting good generalization capabilities, though slightly lower than the training performance.

On the other hand, the Lasso Regression model demonstrated a training R-squared Error of 0.8424 and a test R-squared Error of 0.8710. This model showcased a slightly lower training performance compared to Linear Regression but exhibited better generalization to new data, as evident from the higher R-squared Error on the test set.

In conclusion, both Linear Regression and Lasso Regression models show promise in predicting car prices. The choice between the two depends on the specific requirements and priorities of the application.
